ORDINAL ENCODING FOR main_df

```python
ord_value_selector = ['condition','view','waterfront','grade']
value_subset = main_df[ord_value_selector]
value_subset
```

```python
condition_list = ['Poor', 'Fair', 'Average', 'Good', 'Very Good'] 
view_list = ['NONE','FAIR','AVERAGE','GOOD','EXCELLENT']
waterfront_list = ['NO','YES']
grade_list = ['Low','Low Average','Average','Good','Better','Very Good','Luxury','Excellent','Mansion']
```

```python
o_enc = OrdinalEncoder(categories = [condition_list,view_list,waterfront_list,grade_list])
o_enc.fit(value_subset)
```

```python
X_subset = pd.DataFrame(o_enc.transform(value_subset),
                        columns = value_subset.columns)
X_subset
```
